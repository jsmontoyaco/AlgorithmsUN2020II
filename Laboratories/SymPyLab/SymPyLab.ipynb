1.2 Resolver las integrales de una y tres dimensiones a partir de funciones:


```
def impresion(dx, fx):
  print("la integral de la expresión: ", dx,"dx","\n Es: f(x)= ", fx,"+ C") #función para la impresión de los resultados en las integrales de dos dimensiónes
```

```
def impresionxyz(dxyz, fxyz):
  print("la intregral de la expresión: ", dxyz,"dxdydz", "\n Es: f(x,y,z)= ", fxyz, "+ C") #función para la impresión de los resultados en las integrales de 3 dimensiones
```

```
def integrarxy(dx,x):
  integral= integrate(dx,x)
  return (integral)
```

```
def integrarxyz(dxyz, x, y, z):
  integral= integrate(dxyz, x,y,z)
  return (integral)
```
